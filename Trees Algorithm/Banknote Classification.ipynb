```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../Linear Regression/DATA/data_banknote_authentication.csv')
df.head(10)
```

```python
X = df.drop('Class',axis=1)
y = df['Class']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test , y_train , y_test = train_test_split(X , y , test_size=0.15 , random_state=101)
```

```python
from sklearn.ensemble import RandomForestClassifier
mod = RandomForestClassifier()
mod.fit(x_train , y_train)
from sklearn.model_selection import GridSearchCV
param_grid = { 'n_estimators' : [64 , 100 , 128 , 200] , 'max_features':[2,3,4] , 'bootstrap' : [True, False] , 'oob_score':[True , False]}
grid = GridSearchCV(mod , param_grid)
grid.fit(x_train , y_train)
```

```python
grid.best_estimator_
```

```python
y_pred = grid.predict(x_test)
from sklearn.metrics import classification_report , confusion_matrix
print(classification_report(y_pred , y_test))
print(confusion_matrix(y_pred , y_test))
```

```python
mod2 = RandomForestClassifier()
param_grid2 = {'n_estimators':range(1,201) , 'bootstrap' : [True , False] , 'oob_score' :[True, False] , 'max_features' : [2,3,4]}
grid2 = GridSearchCV(mod2 , param_grid2)
grid2.fit(x_train , y_train)
```
